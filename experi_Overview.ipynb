```python
# 接下来筛出前两位发生停电数目最多的三级区域
df.groupby(df['cityCode']).count().sort_values('poweroffId')
```

```python
# 取前两位的切片吧！；
df_10, df_16 = df[df['cityCode'] == 1340610].copy(), df[df['cityCode'] == 1340616].copy()
```

```python
# 这里看下，用date这个method是可以拿到day步长的
df_10.startTime[2].date()
```

```python
# 看一下数据的分布吧!
df_10.groupby( [t.date() for t in df_10.startTime] ).size().plot(figsize=(20, 7))
```

```python
# 果然，数据量的问题出现了，18——19年七月之间竟然没有任何数据！！这个结果直接使我们的数据少了一半。。。
# 但是没关系，一年的数据量总比没有强，还可以接受。
# 我们接下来只能是把数据集直接截断掉19年大约六月前的数据了。

# 可以猜测，河北省的其他城市也可能是这种情况

# 所以看看另一个吧！
df_16.groupby( [t.date() for t in df_16.startTime] ).size().plot(figsize=(20, 7))
```

```python
# 情况一致，而且借由此还能活发现，数据可能会在1月到3月特别稀少，这就很不利了。
# 我们详细看看16区 一月到三月的数据怎么回事吧

import datetime

df_16[( df_16['startTime'] > np.datetime64('2020-01-01T00:00:00.000000')) & ( df_16['startTime'] < np.datetime64('2020-03-01T00:00:00.000000'))].index
```

```python
# 一月到三月竟然只有不到20条数据，与平均密度相比差了很远；这种情况很不乐观。
# 看一下从每月的数据量吧！ 
df_16.groupby( [t.date().strftime("%Y-%m") for t in df_16.startTime]).size()
```

```python
# 这样看的话，数据缺失的月份应该是 18-5， 18-7 --- 19-5， 20-1， 20-2
# 10号地区也是如此嘛？
df_10.groupby( [t.date().strftime("%Y-%m") for t in df_10.startTime]).size()
```

```python
# 10号地区则不同，只有2月的数据偏少，但也不好说成为缺失。 且10号地区的峰值在10、11月份出现，8月并不高

# 这样的话，我们在选数据上可以借鉴一下此次例子，并不是数据量多就好，还要看均衡水平等。
# 接下来有两个选择，一个是多看几个地方的情况，另一个就是拿一个先做一做啦。

# 这样， 我们再拿两个A3来看看数据情况，保证16号A3不算特殊。

# 先封装之前的过程吧！


def getMonthlyDistr(a2df: pd.DataFrame, a3no: int) -> pd.Series:
    print('a3no:', a3no)
    df = a2df[a2df['cityCode'] == a3no].copy()
    return df.groupby( [t.date().strftime("%Y-%m") for t in df.startTime]).size()
```

```python
print(getMonthlyDistr( df, 13406))
print(getMonthlyDistr( df, 1340612))
print(getMonthlyDistr( df, 1340625))
```
